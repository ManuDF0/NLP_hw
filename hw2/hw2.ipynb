```python
import os
import random
import sys

with open('el_principito.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
words = text.split(' ')
```

```python
# Cuento la cantidad de threegrams
counts = {}
for i in range(0,len(words)-3):
  curr = words[i]
  next = words[i+1]
  following = words[i+2]
  if curr not in counts:
    counts[curr] = {}
  if next not in counts[curr]:
    counts[curr][next] = {}
  if following not in counts[curr][next]:
    counts[curr][next][following] = 1
  else:
    counts[curr][next][following] =+ 1
```

```python
probs = {}
for curr in counts:
  probs[curr] = {}
  for next in counts[curr]:
    probs[curr][next] = {}
    denominator = sum(counts[curr][next].values())
    for following in counts[curr][next]:
      numerator = counts[curr][next][following]
      probs[curr][next][following] = numerator / denominator
```

```python
new_text = []
curr = 'el'
next = 'principito'
new_text.append(curr)
new_text.append(next)

for i in range(len(text)):  # Mismo número de palabras que en el texto original
    if curr in probs and next in probs[curr]:  # Verificación de existencia
        sorted_probs = sorted(probs[curr][next].items(), key=lambda item: item[1], reverse=True)
        new_word = random.choices(population=list(probs[curr][next].keys()), 
                                  weights=list(probs[curr][next].values()), k=1)[0]
        new_text.append(new_word)
        curr = next
        next = new_text[-1]
    else:
        break  # Si no hay una secuencia válida, rompe el bucle

new_text = ' '.join(new_text)
```
