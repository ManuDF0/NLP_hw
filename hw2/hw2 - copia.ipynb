```python
import os
import random
import sys

with open('el_principito.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
words = text.split(' ')
```

```python
def char_model(input, first, second):
    if input.endswith('.txt'):
        # Leer el archivo
        with open(input, 'r', encoding='utf-8') as f:
            text = f.read()
        letters = list(text)

        # Crear un diccionario para contar trigramas
        tri_counts = {}
        for i in range(0, len(letters) - 2):
            first = letters[i]
            second = letters[i + 1]
            curr = letters[i + 2]
            if (first, second) not in tri_counts:
                tri_counts[(first, second)] = {}
            if curr not in tri_counts[(first, second)]:
                tri_counts[(first, second)][curr] = 1
            else:
                tri_counts[(first, second)][curr] += 1

        # Crear un diccionario para las probabilidades
        probs = {}
        for history in tri_counts:
            probs[history] = {}
            total = sum(tri_counts[history].values())  # Total de apariciones para ese historial
            for next_char in tri_counts[history]:
                probs[history][next_char] = tri_counts[history][next_char] / total

        # Generar nuevo texto a partir del historial
        new_text = [first, second]
        history = (first, second)
        for i in range(0, len(text) - 2):
            if history in probs:
                new_letter = random.choices(population=list(probs[history].keys()),
                                            weights=list(probs[history].values()), 
                                            k=1)[0]
                new_text.append(new_letter)
                first = new_text[-2]
                second = new_text[-1]
                history = (first, second)
            else:
                break  # Si no se encuentra el historial, se detiene

        return ''.join(new_text)
    else:
        return 'El input no es un .txt'
```

```python
# Cuento la cantidad de threegrams
counts = {}
for i in range(0,len(words)-3):
  curr = words[i]
  next = words[i+1]
  following = words[i+2]
  if curr not in counts: 
    counts[curr] = {} # Si la palabra actual no está en el diccionario, agrega un key con la palabra actual y como valor le pone un diccionario 
  if next not in counts[curr]:
    counts[curr][next] = {} # Si la palabra que le sigue no está en el diccionario de la palabra actual, le agrega un key con la palabra siguiente y como valor le pone un diccionario
  if following not in counts[curr][next]:
    counts[curr][next][following] = 1 # Si la última palabra del 3-gram no está en el diccionario adentro del diccionario, agrega un key con esa útlima palabra y le da valor 1
  else:
    counts[curr][next][following] =+ 1 # Si la última palabra estaba en el diccionario adentro del diccionario, le suma 1 al valor del key
```

```python
probs = {} # Crea un nuevo diccionario
for curr in counts: # Itera sobre los keys dentro de counts
  probs[curr] = {} # Crea un key dentro de probs con la palabra actual y le pone de valor un diccionario
  for next in counts[curr]: # Itera sobre los keys del diccionario adentro del diccionario de counts
    probs[curr][next] = {} # Agrega keys adentro del diccionario de la palabara actual
    denominator = sum(counts[curr][next].values()) # Suma todos los valores de la cantidad de ocurrencias de ese bigram particular
    for following in counts[curr][next]: # Itera sobre la última palabra de cada 3-gram
      numerator = counts[curr][next][following] # Toma como numerador la cantidad de veces que aparece ese 3-gram en particular
      probs[curr][next][following] = numerator / denominator # Lo divide por la catidad de veces que aparece el bigram previo a esa palabra para obtener la probabilidad condicional de que la tercer palabra sea esa, dadas las dos anteriores
```

```python
new_text = []
curr = 'la'
next = 'flor'
new_text.append(curr)
new_text.append(next)

for i in range(len(text)):  # Mismo número de palabras que en el texto original
    if curr in probs and next in probs[curr]:  # Verificación de existencia
        sorted_probs = sorted(probs[curr][next].items(), key=lambda item: item[0], reverse=True) # Ordena los items del diccionario de acuerdo a la probabilidad y a la primer letra del key
        new_word = random.choices(population=list(probs[curr][next].keys()), 
                                  weights=list(probs[curr][next].values()), k=1)[0] # selecciona aleatoriamente una palabra dentro de las keys del diccionario dentro del diccionario, de acuerdo al peso que tienen
        new_text.append(new_word) # Esa palabra aleatoria se agrega al nuevo texto
        curr = next # Seteo la segunda palabra de mi input como la primer palabra del próximo bigram
        next = new_text[-1] # Seteo la palabra estimada por mi modelo como la segunda palabra del próximo bigram
    else:
        break  # Si no hay una secuencia válida, rompe el bucle

new_text = ' '.join(new_text) # Junto todo el nuevo texto
```
